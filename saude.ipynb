```python
newdata = data.iloc[:, [0, 78, 2, 3, 4, 8, 98, 99, 100, 101, 102]].copy()
newdata.columns = ['Genero', 'Idade', 'Estado Civil', 'Grau Escolar', 'Situação Profiss', 'Orient Política', 'Espontâneo', 'Intuitivo', 'Dependente', 'Evitante', 'Racional']
newdata.head()
```

```python
def classificar_orientacao(valor):
    if valor == 0:
        return 0
    elif 1 <= valor <= 4:
        return 1  # Esquerda
    elif 5 <= valor <= 6:
        return 2
    elif 7 <= valor <= 10:
        return 3
    return 0
```

```python
newdata['Orient Política'] = newdata['Orient Política'].apply(classificar_orientacao)
newdata.head()
```

```python
missing_values = newdata.isnull().sum()
print(missing_values)
```

```python
newdata.iloc[:, 2] = newdata.iloc[:, 2].fillna(0)
newdata.iloc[:, 3] = newdata.iloc[:, 3].fillna(0)
newdata.iloc[:, 4] = newdata.iloc[:, 4].fillna(0)
```

```python
missing_values = newdata.isnull().sum()
print(missing_values)
```

```python
classes = ['Espontâneo', 'Intuitivo', 'Dependente', 'Evitante', 'Racional']
maximos = newdata[classes].max(axis=1)
for classe in classes:
    newdata[classe] = (newdata[classe] == maximos).astype(int)
newdata.head()
```

```python
newdata = newdata.dropna()
missing_values = newdata.isnull().sum()
print(missing_values)
```

```python
newdata.dtypes
```

```python
newdata['Genero'] = newdata['Genero'].astype(int)
newdata['Estado Civil'] = newdata['Estado Civil'].astype(int)
newdata['Grau Escolar'] = newdata['Grau Escolar'].astype(int)
newdata['Situação Profiss'] = newdata['Situação Profiss'].astype(int)
newdata.dtypes
```
